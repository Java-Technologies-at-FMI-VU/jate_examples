```java
void print(Object o) {
    System.out.println(o);
}

void printStackTrace() {
    int skipFirst2 = 0;
    for (StackTraceElement e: Thread.currentThread().getStackTrace()) {
        if (++skipFirst2 > 2) {
            System.out.println(e);
        }
    }
}

void printSuper(Class<?> clazz) {
    if (clazz != null) {
        print(clazz);
        printSuper(clazz.getSuperclass());
    }
}
```

### Catched Exception

```java
public class ExceptionDemo02 {

    void firstCall() throws IOException {
        secondCall();
    }
    
    void secondCall() throws IOException {
        thirdCall();
    }

    void thirdCall() throws IOException {
        throw new IOException();
    }
    
    void firstCall02() {
        try {
            secondCall();
        } catch (IOException ex) {
            ex.printStackTrace();
            printSuper(ex.getClass());
        }
    }
    
}
```

```java
ExceptionDemo02 demo = new ExceptionDemo02();
        
try {

    demo.firstCall();
    
} catch (IOException ex) {

    ex.printStackTrace();
    printSuper(ex.getClass());
    
}
```

```java
demo.firstCall02();
```

### Error

```java
public class ExceptionDemo04 {

    int stackLen = 0;
    String stringLen = "a";
    
    void recursive() {
        stackLen += 1;
        recursive();
    }

    void outOfMemory() {
        while (true) {
            stringLen = stringLen.concat(stringLen);
        }
    }
    
}
```

```java
ExceptionDemo04 demo = new ExceptionDemo04();
        
try {
    demo.recursive();
} catch (Error err) {
    System.err.println(err);
    printSuper(err.getClass());
}

printf("stackLen = %d%n", demo.stackLen);
```

```java
try {
    demo.outOfMemory();
} catch (Error err) {
    System.err.println(err);
    printSuper(err.getClass());
}
        
printf("stringLen = %d = %dMB%n", demo.stringLen.length(), demo.stringLen.length() / 1024 / 1024);
```

### Masking exceptions

```java
interface Type1 {
    
    void f() throws CloneNotSupportedException;
    
}

interface Type2 {
    
    void f() throws InterruptedException;
    
}

interface Type3 {
    
    void f() throws InterruptedException, ArrayIndexOutOfBoundsException;
    
}

interface Type12 extends Type1, Type2 { }

interface Type23 extends Type2, Type3 { }
```

```java
Type12 obj12 = new Type12() {

    @Override
    public void f() {
        print("In f of obj12");
    }

};

void test12() {
    obj12.f();
}

test12();
```

```java
Type23 obj23 = new Type23() {

    @Override
    public void f() {
        System.out.println("In f of obj23");
    }

};


void test23() {
    obj23.f();
}
```

```java
void test23() throws InterruptedException {
    obj23.f();
}
test23();
```
